```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x):
    return 1/x

x = np.linspace(1.001, 10, 100)

theta = np.linspace(0, 2*np.pi, 50)
X, Theta = np.meshgrid(x, theta)

Y = f(X) * np.cos(Theta)
Z = f(X) * np.sin(Theta)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface of Revolution of f(x) = 1/x around the x-axis')

plt.show()
```

```python
# a)

import sympy

x = sympy.Symbol('x')

f_x = 1/x

f_x_squared = f_x**2

volume_integral = sympy.pi * sympy.integrate(f_x_squared, (x, 1, sympy.oo))

volume = volume_integral

print(f"The volume of the shape is: {volume}")
```

```python
# b)

import sympy

x = sympy.Symbol('x')

f_x = 1/x

f_prime_x = sympy.diff(f_x, x)

f_prime_x_squared = f_prime_x**2

integrand = 2 * sympy.pi * f_x * sympy.sqrt(1 + f_prime_x_squared)

antiderivative = sympy.integrate(integrand, x)

print(f"The antiderivative representing the surface area is: {antiderivative}")
```

```python
# c)

import sympy

x = sympy.Symbol('x')
t = sympy.Symbol('t')

f_x = 1/x

f_prime_x = sympy.diff(f_x, x)

f_prime_x_squared = f_prime_x**2

integrand = 2 * sympy.pi * f_x * sympy.sqrt(1 + f_prime_x_squared)

definite_integral = sympy.integrate(integrand, (x, 1, t))

surface_area = sympy.limit(definite_integral, t, sympy.oo)

print(f"The surface area of the shape for 1 < x < infinity is: {surface_area}")
```

# #2

$$g(x) = \frac{x^{19}}{x!}$$

```python
# a)

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

def g(x):
    return (x**19) / gamma(x + 1)

x_values_discrete = np.arange(0, 21, 1)

y_values_discrete = g(x_values_discrete)

plt.figure(figsize=(10, 6))
plt.plot(x_values_discrete, y_values_discrete, 'o', markersize=8) # Use 'o' and increase markersize
plt.xlabel('x (integer values)')
plt.ylabel('g(x)')
plt.title('Graph of g(x) = x^19 / x! for Integer x')
plt.grid(True)
plt.ylim(bottom=0)

plt.show()
```

```python
# b)

import sympy

x = sympy.Symbol('x')

g_x = (x**19) / sympy.gamma(x + 1)

limit_g_x = sympy.limit(g_x, x, sympy.oo)

print(f"The limit of g(x) as x approaches infinity is: {limit_g_x}")
```

```python
# c)

import numpy as np
from scipy.special import gamma

def g(x):
    return (x**19) / gamma(x + 1)

epsilon = 1 / (19**3)
print(f"epsilon is: {epsilon}")

N = None
start_search = 20

for x_val in range(start_search, 100):
    g_at_x = g(x_val)
    if g_at_x < epsilon:
        N = x_val
        break

if N is not None:
    print(f"The smallest integer N such that g(N) < epsilon ({epsilon}) is: {N}")
    if N > 0:
        g_at_N_minus_1 = g(N - 1)
        print(f"For verification, g({N-1}) = {g_at_N_minus_1}")
        if g_at_N_minus_1 < epsilon:
             print(f"Something might be off, g({N-1}) is also less than epsilon. Let's extend the search or re-evaluate the logic.")
else:
    print(f"Could not find an integer N up to 100 such that g(N) < epsilon ({epsilon}). The required N might be larger.")
```

# #3

$$\sum_{n=1}^{\infty} \frac{(\ln n)^2}{n^2}$$

```python
# a)

import sympy

n = sympy.Symbol('n')

a_n = (sympy.ln(n))**2 / (n**2)

N_terms = 1000

partial_sum = sympy.summation(a_n, (n, 1, N_terms))

approx_sum = partial_sum.evalf()

print(f"An approximation of the sum of the series up to {N_terms} terms is: {approx_sum}")
```

$$S_{19} = \sum_{n=1}^{19} \frac{(\ln n)^2}{n^2}$$

```python
# b)

import sympy

x = sympy.Symbol('x')

f_x = (sympy.ln(x))**2 / (x**2)

lower_bound = 19
upper_bound = sympy.oo

error_estimate_integral_19 = sympy.integrate(f_x, (x, lower_bound, upper_bound))

error_estimate_19 = error_estimate_integral_19.evalf()

print(f"The estimated error in using S_19 to approximate the sum is approximately: {error_estimate_19}")
```
